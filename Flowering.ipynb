```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/H2GTRE.TTF" #HY견고딕 폰트
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
a = -1
while a != 0:
    a = int(input('분석하고 싶은 데이터를 선택하세요.\n 1 : 개나리 \n 2 : 매화 \n 3 : 벚꽃 \n 4 : 진달래 \n 0 : 종료 \n'))
    if a == 1:
        temp_f = open('./data/seoul_108_1990_2020.csv')
        data_f = csv.reader(temp_f)
        next(data_f)
        f = open('./data/OBS_계절관측_개나리.csv')
        data= csv.reader(f)
        next(data)
        bloom = []
        temp = []
        for row in data:
            if row[-1] == '개화':
                bloom.append(row[2])
        for row_f in data_f:
            for i in bloom:
                if i == row_f[1]:
                    temp.append(float(row_f[2]))
        bloom_month_day = []
        for k in bloom:
            a = k.split('-')[1]+'-'+k.split('-')[2]
            bloom_month_day.append(a)
        xs = [x for x, _ in enumerate(bloom_month_day)] # bloom_month_day의 index를 저장 0~30 (list 내에서 for문)
        colors = [] #최댓값, 최솟값의 막대만 색을 다르게 하기 위해 빈 list 생성
        for value in temp:
            if value == max(temp):
                colors.append('orangered')
            elif value == min(temp):
                colors.append('royalblue')
            else:
                colors.append('yellow')
        plt.figure(figsize = (30,10))
        plt.bar(xs, temp, width=0.6, color=colors , edgecolor='gray',linewidth=3) #바의 width, edge linewidth설정
        plt.xlabel("1990~2020년") 
        plt.ylabel("평균온도")
        plt.xticks(xs, bloom_month_day) # x좌표 설정(x좌표위치에 들어갈 값)
        plt.tick_params(axis='x',length=5, labelsize=14) #그래프 틱 관련 설정 (길이, 틱 레이블의 크기)
        plt.plot(max(temp), 'orangered', label = '최고 온도') 
        plt.plot(min(temp), 'royalblue', label = '최저 온도')  
        plt.legend()
        plt.title('1990~2020 개나리 개화 시기와 평균 온도')
        
        for index, value in enumerate(temp):
            plt.text(index-0.2, value +0.2, str(value)) # x좌표, y좌표, text
        plt.show()
        f.close()
        temp_f.close()
    elif a == 2:
        temp_f = open('./data/seoul_108_1990_2020.csv')
        data_f = csv.reader(temp_f)
        next(data_f)
        f = open('./data/OBS_계절관측_매화.csv')
        data= csv.reader(f)
        next(data)
        bloom = []
        temp = []
        for row in data:
            if row[-1] == '개화':
                bloom.append(row[2])
        for row_f in data_f:
            for i in bloom:
                if i == row_f[1]:
                    temp.append(float(row_f[2]))
        bloom_month_day = []
        for k in bloom:
            a = k.split('-')[1]+'-'+k.split('-')[2]
            bloom_month_day.append(a)
        xs = [x for x, _ in enumerate(bloom_month_day)]
        colors = []
        for value in temp:
            if value == max(temp):
                colors.append('orangered')
            elif value == min(temp):
                colors.append('royalblue')
            else:
                colors.append('lightgoldenrodyellow')
        plt.figure(figsize = (30,10))
        plt.bar(xs, temp, width=0.6, color=colors , edgecolor='gray',linewidth=3)
        plt.xlabel("1990~2020년")
        plt.ylabel("평균온도")
        plt.xticks(xs, bloom_month_day)
        plt.tick_params(axis='x',length=5, labelsize=14)
        plt.plot(max(temp), 'orangered', label = '최고 온도') 
        plt.plot(min(temp), 'royalblue', label = '최저 온도')  
        plt.legend()
        plt.title('1990~2020 매화 개화 시기와 평균 온도')
        
        for index, value in enumerate(temp):
            plt.text(index-0.2, value +0.2, str(value))
        plt.show()
        f.close()
        temp_f.close()
    elif a == 3:
        temp_f = open('./data/seoul_108_1990_2020.csv')
        data_f = csv.reader(temp_f)
        next(data_f)
        f = open('./data/OBS_계절관측_벚꽃.csv')
        data= csv.reader(f)
        next(data)
        bloom = []
        temp = []
        for row in data:
            if row[-1] == '개화':
                bloom.append(row[2])
        for row_f in data_f:
            for i in bloom:
                if i == row_f[1]:
                    temp.append(float(row_f[2]))
        bloom_month_day = []
        for k in bloom:
            a = k.split('-')[1]+'-'+k.split('-')[2]
            bloom_month_day.append(a)
        xs = [x for x, _ in enumerate(bloom_month_day)]
        colors = []
        for value in temp:
            if value == max(temp):
                colors.append('orangered')
            elif value == min(temp):
                colors.append('royalblue')
            else:
                colors.append('lavenderblush')
        plt.figure(figsize = (30,10))
        plt.bar(xs, temp, width=0.6, color=colors , edgecolor='gray',linewidth=3)
        plt.xlabel("1990~2020년")
        plt.ylabel("평균온도")
        plt.xticks(xs, bloom_month_day)
        plt.tick_params(axis='x',length=5, labelsize=14)
        plt.plot(max(temp), 'orangered', label = '최고 온도') 
        plt.plot(min(temp), 'royalblue', label = '최저 온도')  
        plt.legend()
        plt.title('1990~2020 벚꽃 개화 시기와 평균 온도')
        
        for index, value in enumerate(temp):
            plt.text(index-0.2, value +0.2, str(value))
        plt.show()
        f.close()
        temp_f.close()
    elif a == 4:
        temp_f = open('./data/seoul_108_1990_2020.csv')
        data_f = csv.reader(temp_f)
        next(data_f)
        f = open('./data/OBS_계절관측_진달래.csv')
        data= csv.reader(f)
        next(data)
        bloom = []
        temp = []
        for row in data:
            if row[-1] == '개화':
                bloom.append(row[2])
        for row_f in data_f:
            for i in bloom:
                if i == row_f[1]:
                    temp.append(float(row_f[2]))
        bloom_month_day = []
        for k in bloom:
            a = k.split('-')[1]+'-'+k.split('-')[2]
            bloom_month_day.append(a)
        xs = [x for x, _ in enumerate(bloom_month_day)]
        colors = []
        for value in temp:
            if value == max(temp):
                colors.append('orangered')
            elif value == min(temp):
                colors.append('royalblue')
            else:
                colors.append('orchid')
        plt.figure(figsize = (30,10))
        plt.bar(xs, temp, width=0.6, color=colors , edgecolor='gray',linewidth=3)
        plt.xlabel("1990~2020년")
        plt.ylabel("평균온도")
        plt.xticks(xs, bloom_month_day)
        plt.tick_params(axis='x',length=5, labelsize=14)
        plt.plot(max(temp), 'orangered', label = '최고 온도') 
        plt.plot(min(temp), 'royalblue', label = '최저 온도')  
        plt.legend()
        plt.title('1990~2020 진달래 개화 시기와 평균 온도')
        
        for index, value in enumerate(temp):
            plt.text(index-0.2, value +0.2, str(value))
        plt.show()
        f.close()
        temp_f.close()
    elif a == 0:
        print('종료합니다')
        break
    else:
        print('잘못된 선택입니다. 다시 시도 해주세요.')
```
